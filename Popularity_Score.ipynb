```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url='https://insights.blackcoffer.com/rising-it-cities-and-its-impact-on-the-economy-environment-infrastructure-and-city-life-by-the-year-2040-2/'
response=requests.get(url)
response
```

```python
soup=BeautifulSoup(response.content,'html.parser')
```

```python
text=soup.find_all('div',attrs={'class':'td-post-content tagdiv-type'})
```

```python
text_data=[]
for i in text:
    txt=i.text
    text_data.append(txt)
```

```python
text_data
```

```python
text_str=' '.join([str(elem) for elem in text_data])
```

```python
text_str
```

```python
words_lis=text_str.split()
```

```python
words_lis
```

```python
positive_words_list=pd.read_table('positive-words.txt', header=None, delimiter=None)[0].to_list()
```

```python
positive_words_list
```

```python
count=0
for i in words_lis:
    for j in positive_words_list:
        if i[:]==j[:]:
            count+=1
```

```python
print(count)
```

```python
pos_percent=count/len(positive_words_list)
```

```python
pos_percent
```

```python
negative_words_list=pd.read_table('negative-words.txt', header=None, delimiter=None,encoding='latin-1')[0].to_list()
```

```python
negative_words_list
```

```python
neg_count=0

for i in negative_words_list:
    for j in words_lis:
        if i==j:
            neg_count+=1
```

```python
neg_per=neg_count/len(negative_words_list)
```

```python
neg_per
```

```python
popularity_score=(pos_percent-neg_per)/((pos_percent+neg_per)+0.000001)
```

```python
popularity_score
```
